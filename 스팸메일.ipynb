```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
porter = PorterStemmer()
lemma = WordNetLemmatizer()
```

```python
world_dic = []
text = []

def make_dic(data, dic):
    for i in range(0, 5559):
        for j in range(0,len(data['text'][i].split(' '))):
            dic.append(data['text'][i].split(' ')[j])
    for i in range(0,len(dic)):
        dic[i] = porter.stem(dic[i])                  
              
make_dic(spam, world_dic)                   
```

```python
for i in range(0,len(world_dic)):
    world_dic[i] = lemma(world_dic[i])
```
