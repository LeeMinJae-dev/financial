```python
import pandas as pd

train_o = pd.read_csv('train-data.csv')
test_o = pd.read_csv('test-data.csv')

```

```python
train = train_o.copy()
test = test_o.copy()

train.drop('Unnamed: 0', axis = 1,inplace= True)
test.drop('Unnamed: 0', axis = 1,inplace= True)

train['Seats'].value_counts()
```

```python
train['Seats'] = train['Seats'].fillna(5)
train['Seats'] = train['Seats'].astype(int)

test['Seats'] = test['Seats'].fillna(5)
test['Seats'] = test['Seats'].astype(int)
```

```python
train['Name'] = train['Name'].str.extract('([A-Za-z]+)')
test['Name'] = test['Name'].str.extract('([A-Za-z]+)')
```

```python
train.loc[4446,'Mileage'] = 29.3
train['Mileage'].dropna()
```

```python
train['Mileage'] = train["Mileage"].str.extract('([0-9]+\.[0-9]+)')
test['Mileage'] = test["Mileage"].str.extract('([0-9]+\.[0-9]+)')
```

```python
train['Mileage'] = train['Mileage'].astype(float)
test['Mileage'] = test['Mileage'].astype(float)
```

```python
train.loc[train['Mileage'] == 0, 'Mileage'] = train['Mileage'].median()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy= 'mean')
```

```python
train['Power'] = train["Power"].str.extract('([0-9]+\.[0-9]+)')
test['Power'] = test["Power"].str.extract('([0-9]+\.[0-9]+)')
```

```python
train.replace({"First":1,"Second":2,"Third": 3,"Fourth & Above":4},inplace=True)
test.replace({"First":1,"Second":2,"Third": 3,"Fourth & Above":4},inplace=True)

train['Owner_Type']
```
